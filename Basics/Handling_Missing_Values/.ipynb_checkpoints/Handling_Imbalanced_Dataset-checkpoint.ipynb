### Imbalanced Dataset:

A dataset with an unequal class distribution.

```python
# importing the dependencies
import numpy as np
import pandas as pd
```

```python
# loading the dataset to pandas DataFrame
credit_card_data = pd.read_csv('credit_data.csv')
```

```python
# first 5 rows of the dataframe
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# distribution of the two classes
credit_card_data['Class'].value_counts()
```

0 --> Legit Transaction

1 --> Fradulent Transaction

This is Highly Imbalanced Dataset

```python
# separating the legit and fradulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

### Under-sampling

Building a sample dataset containing similar distribution of Legit & Fradulent Transactions

Number of Fradulent Transactions-->492

```python
legit_sample = legit.sample(n=492)  # random 492 samples will be selected.
```

```python
print(legit_sample.shape)
```

### Concatenate the Two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0) 
# here axis = 0 means we want to concatenate the DataFrames one top of the other or row wise..  
```
